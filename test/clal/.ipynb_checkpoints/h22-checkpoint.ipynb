```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

fig_width_pt = 3*246.0  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inch
golden_mean = (np.sqrt(5)-1.0)/2.0         # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig_height = fig_width*golden_mean      # height in inches
fig_size =  [fig_width,fig_height]

params = { 'axes.labelsize': 24,
          'font.family': 'serif',
          'font.serif': 'Computer Modern Raman',
          'font.size': 24,
          'legend.fontsize': 20,
          'xtick.labelsize': 24,
          'ytick.labelsize': 24,
          'axes.grid' : True,
          'text.usetex': True,
          'savefig.dpi' : 100,
          'lines.markersize' : 14,
          'figure.figsize': fig_size}

mpl.rcParams.update(params)
```

```python
import lal
import lalsimulation as lalsim
import numpy as np
```

```python
approx=lalsim.IMRPhenomPv2


m1 = 30
m2 = 30

s1 = [0,0,0]#[0.4,-0.2,0.43]
s2 = [0,0,0]#[-0.1,0.8,0]
dist = 400.
iota = 0
phi_c = 0.

deltaT = 1./4096
f_ref = 20.
f_low = 40

longAscNodes = 0
eccentricity = 0 
meanPerAno = 0
nonGRdict = lal.CreateDict()
```

```python
hpt1, hct1 = lalsim.SimInspiralChooseTDWaveform(m1 * lal.MSUN_SI, m2 * lal.MSUN_SI,\
                                                          s1[0], s1[1], s1[2],\
                                                          s2[0], s2[1], s2[2],\
                                                          dist * 1e6 * lal.PC_SI, iota, phi_c,\
                                                          longAscNodes, eccentricity, meanPerAno,
                                                          deltaT, f_low, f_ref,
                                                          nonGRdict, approx)

t1 = np.arange(hpt1.data.length, dtype=float) * hpt1.deltaT
t1 = t1 + hpt1.epoch
plt.plot(t1, hpt1.data.data, label='hplus:SEOBNRv4,lal')
plt.xlabel('Time [s]')
plt.ylabel('Strain')
```
